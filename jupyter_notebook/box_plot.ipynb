```python
import seaborn as sns

# make canvas white grid

sns.set(style="whitegrid")
tips= sns.load_dataset("tips")

sns.boxplot(x="tip",y="day",data=tips,hue="smoker",palette="set2" ,dodge=False)
```

```python
# orientation of the plot x and y axis
import seaborn as sns

# make canvas white grid

sns.set(style="whitegrid")
tips= sns.load_dataset("tips")

sns.boxplot(x="tip",y="day",data=tips,hue="smoker",orient="h",dodge=False)
```

```python
# orientation of the plot x and y axis
import seaborn as sns

# make canvas white grid

sns.set(style="whitegrid")
tips= sns.load_dataset("tips")

sns.boxplot(x="tip",y="day",data=tips,hue="smoker",orient="h",dodge=True) #dodge mean passing by
# if you need to give different color to each bar you must use hue funciton
```

```python
seaborn.boxplot(x="survived",
                y="age",showmeans=True,
                data= kashti)
```

```python
seaborn.boxplot(x="survived",
                y="age",showmeans=True,meanprops={"marker":"*","markersize":"12", "markeredgecolor":"red"},
                data= kashti)
```

```python
import matplotlib.pyplot as plt
seaborn.boxplot(x="survived",
                y="age",showmeans=True,meanprops={"marker":"*","markersize":"12", "markeredgecolor":"red"},
                data= kashti)
plt.xlabel("survived",size=18)
plt.ylabel("age (years)",size=14)
plt.title("kitna dobay kitna bucha",size=20,weight="bold")
plt.show()
#show labels

```
