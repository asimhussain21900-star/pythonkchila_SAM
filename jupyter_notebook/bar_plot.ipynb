```python
# To change color SINCE WARNING IN FUTURE COLOR CAN BE USED AS PALETTE
import seaborn as sns
import matplotlib.pyplot as plt

# load data set

kashti=sns.load_dataset("titanic")
kashti
# draw a bar plot
sns.barplot(x="sex",y="alone",hue="who", data=kashti,order=["female","male"], palette='dark:red')
plt.show()
```

```python
# To change CANDLE INDENT 
import seaborn as sns
import matplotlib.pyplot as plt

# load data set

kashti=sns.load_dataset("titanic")
kashti
# draw a bar plot
sns.barplot(x="sex",y="alone",hue="who", data=kashti,order=["female","male"], palette='dark:red',ci=None)
plt.show()
```

```python
# To change CANDLE INDENT since warning ci can be replaced by errorbar
import seaborn as sns
import matplotlib.pyplot as plt

# load data set

kashti=sns.load_dataset("titanic")
kashti
# draw a bar plot
sns.barplot(x="sex",y="alone",hue="who", data=kashti,order=["female","male"], palette='pastel',errorbar=None)
plt.show()
```

```python
# to add an estimater but we need to add numpy library to import mean mode and median
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean
from numpy import median
import numpy
# load data set

kashti=sns.load_dataset("titanic")
kashti
# draw a bar plot
sns.barplot(x="class",y="fare",hue="sex", data=kashti, palette='pastel',estimator=median)
plt.show()
```

```python
# to increase or decrease color saturation in plot
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean
from numpy import median
import numpy
# load data set

kashti=sns.load_dataset("titanic")
kashti
# draw a bar plot
sns.barplot(x="class",y="fare",hue="sex", data=kashti, palette='pastel',estimator=median,saturation=0.1)
plt.show()
```

```python
#Horizontal plot to make the bar horizontal simply change x axis values to numeric
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean
from numpy import median
import numpy
# load data set

kashti=sns.load_dataset("titanic")
kashti
# draw a bar plot
sns.barplot(x="fare",y="class",hue="sex", data=kashti, palette='pastel',estimator=median,saturation=0.1)
plt.show()
```

```python
# To play with color and width of the bar lines
import seaborn as sns
import matplotlib.pyplot as plt

# load data set

kashti=sns.load_dataset("titanic")
kashti
# draw a bar plot
sns.barplot(x="class",y="fare", data=kashti,
            linewidth=2.5, facecolor=(1,1,1,0),
            errcolor="0.2", edgecolor="0.2")
plt.show()
```

```python
# To play with color and width of the bar lines to be used in future version and removing above failed line
import seaborn as sns
import matplotlib.pyplot as plt

# load data set

kashti=sns.load_dataset("titanic")
kashti
# draw a bar plot / # values of facecolor err_kws and edge color will be between 0 to 1
sns.barplot(x="class",y="fare", data=kashti,
            linewidth=2.5, facecolor=(1,1,1,0),
            err_kws={"color":"0.2"}, edgecolor="0.2")
plt.show()
```
